```python
import random
class LLMdup():
    def __init__(self):
        print("LLM created")

    def predict(self, prompt):
        response_list=[
            "Dummy response 1",
            "Dummy response 2",
            "Dummy response 3"
        ]
        return {"Response":random.choice(response_list) + '\nFor Prompt: ' + prompt}
```

```python
llm = LLMdup()
```

```python
class PromptTemplatedup():
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = PromptTemplatedup(
    template="Write a {length} poem on {topic}.",
    input_variables=["length", "topic"]
)

prompt = template.format({"legnth": "short", "topic":"love"})
print(prompt)
```

```python
result = llm.predict(prompt)
print(result)
```

```python
class LLMChaindup():
    def __init__(self, llm, prompt):
        self.llm = llm
        self.prompt = prompt
        
    def run(self, input_dict):
        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)
        return result
    
```

```python
template = PromptTemplatedup(
    template="Write a {length} poem on {topic}.",
    input_variables=["length", "topic"]
)
```

```python
chain=LLMChaindup(llm=llm, prompt=template)
result_chain=chain.run({"length":"short", "topic":"friendship"})
print(result_chain)
```
